```python
import pandas as pd
from pandas_datareader import data

import matplotlib.pyplot as plt
%matplotlib inline

fname = "/root/Desktop/WikiPageRanking/WikiPageRanking/output/part-00000"

with open(fname, "r") as ins:
    array = []
    for line in ins:
        array.append(line)
        
links = {}
for i in range(0,10): 
    title = array[i].split('>')
    credit = title[1][0:-2]
    links[title[0]] = float(credit)
    
ser = pd.Series(links)

ser.plot(kind = "barh")
```
